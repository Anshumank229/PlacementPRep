Finding Factorial of a number 

```python
n=int(input("Enter the number for which you want factorial "))
factorial=1
while n>0:
    factorial=factorial*n
    n=n-1
factorial
```

Finonacci Series

```python
n=int(input("Enter the number "))
a,b=100,101
counter=0
while counter<n:
    print(a)
    c=a+b
    a=b
    b=c
    counter=counter+1
```
